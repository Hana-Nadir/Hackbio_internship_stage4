```python
# STEP 1: importing the docking result csv file
```

```python
import pandas as pd
df_HDAC1 = pd.read_csv('HDAC1 docking result.csv')
df_HDAC1
```

```python
# STEP 2: calculating the mean and std
df_HDAC1['Mean'] = df_HDAC1[['Binding Affinity 1', 'Binding Affinity 2', 'Binding Affinity 3']].mean(axis=1)
df_HDAC1['std'] = df_HDAC1[['Binding Affinity 1', 'Binding Affinity 2', 'Binding Affinity 3']].std(axis=1)
```

```python
df_HDAC1
```

```python
#STEP 4: order according to mean value and livng only models with highest affinity from each ligand
```

```python
df_HDAC1 = df_HDAC1.sort_values(by=['Ligand', 'Mean'])
df_HDAC1 = df_HDAC1.drop_duplicates(subset=['Ligand'])
df_HDAC1 = df_HDAC1.sort_values(by=['Mean'])
df_HDAC1
```

```python
#  STEP 3: genarting the heat map
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

sns.heatmap(df_HDAC1[['Mean']], cmap='viridis')

ax = sns.heatmap(df_HDAC1[['Mean']])
ax.collections[0].colorbar.set_label("Mean")

plt.title("Heatmap of Mean of The Binding Affinity")
sns.heatmap(df_HDAC1[['Mean']])
plt.show()
```

```python
#  STEP 5: saving file as csv
```

```python
df_HDAC1.to_csv('df_HDAC1_II.csv', index=False)
```

# Creating a heatmap to visualize the binding affinities of compounds (with PubChem CIDs on the y-axis) across 11 HDAC subtypes (on the x-axis).

```python
import os
import pandas as pd

#Directory containing the Excel files for all HDACs with all ligands and the mean binding affinities for each HDAC subtype 
directory = '/home/bosibori/sept_hackbio_internship/Stage4/Docking_results/Ligand_mean'

print(f"Checking directory: {directory}")
```

```python
print("Files in the directory:")
for filename in os.listdir(directory):
    print(filename)
```

```python
# List to store DataFrames
dfs = []

# Read all Excel files
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(directory, filename)
        print(f"\nProcessing file: {file_path}")
        try:
            df = pd.read_excel(file_path, names=['PubChem CID', 'Mean'], engine='openpyxl')
            df.set_index('PubChem CID', inplace=True)
            df.columns = [filename[:-5]]  # Use filename (without .xlsx) as column name
            dfs.append(df)
            print(f"Successfully read {filename}. Shape: {df.shape}")
            print(f"First few rows of {filename}:")
            print(df.head())
        except Exception as e:
            print(f"Error reading {filename}: {str(e)}")

print(f"\nTotal DataFrames created: {len(dfs)}")

if len(dfs) == 0:
    print("No Excel files were successfully read. Please check the file contents.")
else:
    try:
        # Combine all DataFrames
        merged_df = pd.concat(dfs, axis=1)

        # Sort each column independently
        for column in merged_df.columns:
            merged_df[column] = merged_df[column].sort_values(ascending=False)

        # Reset index to make PubChem CID a column
        merged_df = merged_df.reset_index()

        # Reorder columns to have PubChem CID first
        columns = ['PubChem CID'] + [col for col in merged_df.columns if col != 'PubChem CID']
        merged_df = merged_df[columns]

        # Save the result to a new Excel file
        output_file = os.path.join(directory, 'merged_hdac_data.xlsx')
        merged_df.to_excel(output_file, index=False, engine='openpyxl')

        print(f"\nMerged data saved to '{output_file}'")
        print(f"Final DataFrame shape: {merged_df.shape}")
        print("\nFirst few rows of the merged data:")
        print(merged_df.head())
    except Exception as e:
        print(f"\nError during merging or saving: {str(e)}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

file_path = '/home/bosibori/sept_hackbio_internship/Stage4/Docking_results/Ligand_mean/merged_hdac_data.xlsx'
df = pd.read_excel(file_path)

# Set 'PubChem CID' as the index
df.set_index('PubChem CID', inplace=True)

# Replace any potential infinity or very large values with NaN
df = df.replace([np.inf, -np.inf], np.nan)

# Drop rows with any NaN values
df = df.dropna()
```

```python
# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df, annot=False, cmap='viridis', center=0)

plt.title('HDAC Binding Affinity Heatmap')
plt.xlabel('HDAC Subtypes')
plt.ylabel('PubChem CID')
```

```python
# Save the heatmap
output_file = '/home/bosibori/sept_hackbio_internship/Stage4/Docking_results/Ligand_mean/hdac_heatmap.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')

print(f"Heatmap saved to: {output_file}")
```
